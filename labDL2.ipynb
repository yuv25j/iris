```python
import numpy as np

x =np.array([1,5,3,8,6,7,2,8,2,1])
y =np.array([20,11,17,15,14,13,18,19,13,2])
data = list(zip(x,y))
data
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
model = KMeans(n_clusters =2)
model.fit(data)
plt.scatter(x,y,c = model.labels_)
plt.show
```

```python
import pandas as pd
dataa = pd.read_csv("D:/iris.csv")
dataa
x=dataa['sepal.length']
y=dataa['petal.length']
dataa=list(zip(x.values , y.values))
dataa
```
